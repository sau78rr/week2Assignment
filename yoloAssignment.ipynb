```python
import os
import json
import matplotlib.pyplot as plt
from ultralytics import YOLO
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Define a function to extract true labels from filenames
def extract_true_labels(test_data_path):
    true_labels = []
    for filename in os.listdir(test_data_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            true_label = 'no_damage' if 'no_damage' in filename else 'damage'
            true_labels.append(true_label)
    return true_labels

# Define a function to extract predicted labels from the results object
def extract_predicted_labels(results):
    predicted_labels = []
    for result in results:
        if result.boxes:
            # Extract the class indices from the result
            class_indices = result.boxes.cls
            # Map class indices to class names using the 'names' dictionary
            class_names = [result.names[int(idx)] for idx in class_indices]
            predicted_labels.extend(class_names)
        else:
            # If no detections, assume 'no_damage'
            predicted_labels.append('no_damage')
    return predicted_labels

# Load the YOLO model
model = YOLO('yolov8n.yaml')  # Replace with the actual path to your model

# Perform predictions
test_data_path = '/content/test-20240712T174414Z-001/test/images'
results = model.predict(source=test_data_path, save=True)  # Save results for later visualization

# Extract true and predicted labels
true_labels = extract_true_labels(test_data_path)
predicted_labels = extract_predicted_labels(results)

# Print the labels for debugging
print(f"True Labels: {true_labels}")
print(f"Predicted Labels: {predicted_labels}")

# Ensure the lengths match before proceeding to generate the confusion matrix
if len(true_labels) == len(predicted_labels):
    # Generate confusion matrix
    cm = confusion_matrix(true_labels, predicted_labels, labels=['damage', 'no_damage'])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['damage', 'no_damage'])

    # Plot and save the confusion matrix
    fig, ax = plt.subplots(figsize=(10, 7))
    disp.plot(ax=ax)
    plt.savefig('confusion_matrix.png')
    plt.show()
else:
    print("Error: Number of true labels does not match number of predicted labels.")

# Generate and save results.png
def save_results_visualization(results, save_path):
    # Create a blank figure
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set_title('Detection Results Summary')

    # Display results for the first few images
    num_images = min(9, len(results))
    for i in range(num_images):
        result = results[i]
        img = result.orig_img
        ax = plt.subplot(3, 3, i + 1)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(f"Image {i+1}")

    # Save the results visualization
    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()

# Save the results.png visualization
save_results_visualization(results, 'runs/detect/train/results.png')
```
